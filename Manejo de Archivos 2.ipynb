```python
path = "/home/lcampos/zeus/proyecto_fondecyt/VSD_4ms2/charmm-gui-2284521356/gromacs/energy_xvg/"
dir_list = os.listdir(path)

archivos_xvg = [archivo for archivo in dir_list if archivo.endswith('.xvg')]

# Lista de nombres de archivos
nombres_de_archivos = archivos_xvg

# Expresión regular para buscar el valor de "E="
patron_e = r"E=([-]?\d+,\d+)"

# Expresión regular para buscar el carácter 'R' seguido de un número
patron_r = r"_R(\d+)_"

patron_ener = r"energy_(\d+)"
```

```python
# Crear un diccionario para almacenar los grupos
arreglo = {}

# Filtrar y agrupar los archivos en el diccionario
for archivo in nombres_de_archivos:
    nombre_archivo = path + archivo
    coincidencias_e = re.search(patron_e, nombre_archivo)
    coincidencias_r = re.search(patron_r, nombre_archivo)
    coincidencias_ener = re.search(patron_ener, nombre_archivo)
    
    if coincidencias_e and coincidencias_r and coincidencias_ener:
        grupo = coincidencias_e.group(1)
        rep = coincidencias_r.group(1)
        energy = coincidencias_ener.group(1) 
        
        try:
            # Abre el archivo en modo lectura
            with open(nombre_archivo, 'r') as archivo:
                # Ignorar las líneas de encabezado que comienzan con '@'
                for linea in archivo:
                    if not linea.startswith('@') and not linea.startswith('#'):
                        partes = linea.strip().split()
                        if len(partes) == 2:
                            x, y = partes
                            x, y = float(x), float(y)

                            if grupo not in arreglo:
                                arreglo[grupo] = {}

                            if rep not in arreglo[grupo]:
                                arreglo[grupo][rep] = {}

                            if x not in arreglo[grupo][rep]:
                                arreglo[grupo][rep][x] = {}

                            arreglo[grupo][rep][x][energy] = y
        except FileNotFoundError:
            print(f"El archivo '{nombre_archivo}' no se encuentra.")
        except Exception as e:
            print(f"Se produjo un error al abrir o procesar el archivo: {e}")
```
